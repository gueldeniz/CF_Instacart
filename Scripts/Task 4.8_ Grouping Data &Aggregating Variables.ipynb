### 4) Create a loyalty flag for existing customers using the transform() and loc() functions.

#### Aggregating data

```python
# Split the data into groups based on the “user_id” column.=> df_ords_prods_merged.groupby(['user_id'])
## Apply the transform() function on the “order_number” column to generate the maximum orders for each user.=> ['order_number'].transform(np.max)
### Create a new column, “max_order,” into which you’ll place the results of your aggregation.=> df_ords_prods_merged['max_order'

ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head(15)
```

#### Deriving columns with loc()

```python
# If the maximum orders the user has made is over 40, then the customer will be labeled a “Loyal customer.”

ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
# If the maximum orders the user has made is over 10 but less than or equal to 40, then the customer will be labeled a “Regular customer.”

ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
# If the maximum orders the user has made is less than or equal to 10, then the customer will be labeled a “New customer.”

ords_prods_merge.loc[ords_prods_merge['max_order'] <=10, 'loyalty_flag'] = 'New customer'
```

### 6) Create a spending flag for each user based on the average price across all their orders using the following criteria:
##### a) If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
##### b) If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
ords_prods_merge.groupby('user_id').agg({'prices':'mean'}).head()
```

```python
# Split the data into groups based on the “user_id” column.
## Apply the transform() function on the “prices” column to generate the average orders for each user.
### Create a new column, “avg_price,” into which you’ll place the results of your aggregation.

ords_prods_merge['avg_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head()
```

#### Deriving columns with loc()

```python
# a) Creating 'Low spender' spending flag for customers whose average spending is lower than 10.

ords_prods_merge.loc[ords_prods_merge['avg_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
# b) Creating 'High spender' spending flag for customers whose average spending is higher than or equal to 10.

ords_prods_merge.loc[ords_prods_merge['avg_price'] >= 10, 'spending_flag'] = 'High spender'
```

### 7) Determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:
##### a) If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
##### b) If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
##### c) If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
# Split the data into groups based on the “user_id” column.
## Apply the transform() function on the “days_since_prior_order” column to generate the median orders for each user.
### Create a new column, “order_frequency,” into which the results of your aggregation will be placed.

ords_prods_merge['order_regularity'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

#### Deriving columns with loc()

```python
# a) Creating a frequency flag called 'Non-frequent customer' for customers whose median of "days_since_last_order" is higher than 20.

ords_prods_merge.loc[ords_prods_merge['order_regularity'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
# b) Creating a frequency flag called 'Regular customer' for customers whose median of "days_since_last_order" is higher than 10 or equal to 20.


ords_prods_merge.loc[(ords_prods_merge['order_regularity'] <= 20) & (ords_prods_merge['order_regularity'] > 10 ), 'frequency_flag'] = 'Regular customer'
```

```python
# a) Creating a frequency flag called 'Frequent customer' for customers whose median of "days_since_last_order" is lower or equal to 10.

ords_prods_merge.loc[ords_prods_merge['order_regularity'] <= 10, 'frequency_flag'] = 'Frequent customer'
```
